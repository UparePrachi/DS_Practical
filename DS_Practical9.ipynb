```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('titanic.csv')
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Age']=data['Age'].fillna(data['Age'].mode()[0])
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

# boxplot

```python
sns.boxplot(data['Sex'],data["Age"], hue=data["Survived"])
```

```python
sns.boxplot(data['Pclass'], data["Age"],hue= data["Survived"])
```

```python
sns.boxplot(data['SibSp'], data["Age"], hue=data["Survived"])
```

```python
sns.boxplot(data['Parch'],data['Age'],hue=data['Survived'])
```

```python
sns.boxplot(data['Embarked'],data['Age'],hue=data['Survived'])
```

# Stripplot

```python
sns.stripplot(data['Age'],data['Sex'])
```

```python
sns.stripplot(data['Age'],data['Sex'],hue=data['Survived'])
```

```python
sns.stripplot(data['Age'],data['Sex'],hue=data['Survived'],split=True)
```

```python
sns.stripplot(data['Pclass'],data['Age'],hue=data['Survived'])
```
