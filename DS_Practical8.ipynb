```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('titanic.csv')
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(np.mean(data['Age']))
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

# Countplot

```python
sns.countplot(data['Sex'],hue=data['Survived'])
```

```python
sns.countplot(data['Survived'])
```

```python
sns.countplot(data['Pclass'])
```

```python
sns.countplot(data['Embarked'])
```

# Violinplot

```python
sns.violinplot(data['Sex'],data['Age'],hue=data['Survived'])
```

```python
sns.violinplot(data['Sex'],data['Age'],hue=data['Survived'],split=True)
```

# Jointplot

```python
sns.jointplot(data['Fare'],data['Age'])
```

```python
sns.jointplot(data=data,x="Survived", y="Fare")
```

# Catplot

```python
sns.catplot(data=data, x="Age", y="Sex",hue='Survived')
```

```python
sns.catplot(data=data, x="Age", y="Sex",hue='Survived',split=True)
```

```python
sns.catplot(data=data,x= 'Pclass', y = 'Age', kind = 'box')
```

```python
sns.catplot(x= 'Pclass', y = 'Fare', data=data, kind = 'strip')
```

```python
sns.catplot(x= 'Sex', y = 'Fare', data=data, kind = 'strip')
```

# Factorplot

```python
sns.factorplot(data=data,x='Fare',y='Age',hue='Survived')
```

# HeatMap

```python
datacorr=data.corr()
datacorr
```

```python
sns.heatmap(datacorr,annot=True)
```

# paitplot

```python
sns.pairplot(data)
```

```python
sns.pairplot(data,hue='Sex')
```

# distplot

```python
sns.distplot(data['Fare'])
```

```python
sns.distplot(data['Fare'],kde=False)
```

```python
sns.distplot(data['Age'])
```
