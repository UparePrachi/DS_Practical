Create a Linear Regression Model using Python/R to predict home prices using Boston Housing Dataset (https://www.kaggle.com/c/boston-housing). The Boston Housing dataset contains 
information about various houses in Boston through different parameters. There are 506 samples and 13 feature variables in this dataset.

The objective is to predict the value of prices of the house using the given features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
data=pd.read_csv("boston_housing.csv")
data
```

```python
data.describe()
```

```python
x=data['lstat']
y=data['medv']
plt.scatter(x,y)
```

```python
x=data['rm']
y=data['medv']
plt.scatter(x,y)
```

```python
x=data[['lstat','rm',"crim",'zn','indus','chas','nox','age','dis','rad','tax','ptratio','black']]
#x=data[['lstat','rm']]
y=data['medv']
```

## Train Test Split Model 

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
#size of the models
print("Size of X_train:-",x_train.size)
print("Size of X_test:-",x_test.size)
print("Size of y_train:-",y_train.size)
print("Size of y_test:-",y_test.size)
```

```python
#Shape of the models
print("Shape of X_train:-",x_train.shape)
print("Shape of X_test:-",x_test.shape)
print("Shape of y_train:-",y_train.shape)
print("Shape of y_test:-",y_test.shape)
```

## Linear Regression 

```python
#create the object of linear regression
lr=LinearRegression()
```

```python
#fit the train models 
lr.fit(x_train,y_train)
```

```python
print(lr.coef_)
```

```python
y_pred=lr.predict(x_test)
```

```python
pd.DataFrame(np.c_[x_test,y_test,y_pred],columns=["crim",'zn','indus','chas','rm','nox','age','dis','rad','tax','ptratio','black','lstat','medv',"medv_pred"])
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
print(np.mean((y_pred - y_test)**2))
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
mse = metrics.mean_squared_error(y_test,y_pred)
mae = metrics.mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
print(f'Mean Squared Error: {mse}',f'Mean Absolute Error: {mae}',f'Root Mean Squared Error: {rmse}',sep='\n')
```
